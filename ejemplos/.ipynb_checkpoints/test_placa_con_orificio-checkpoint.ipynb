# Test - Placa con orificio

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from femppp import * 
```

## Problema 1

- Empotrado a la izquierda
- Carga distribuida a la derecha en 0 grados

```python

def generate_plate_with_hole(L, H, D, nx, ny, n_theta=40):
    center_x = L/2
    center_y = H/2
    radius = D/2

    x = np.linspace(0, L, nx)
    y = np.linspace(0, H, ny)
    X, Y = np.meshgrid(x, y)
    points_grid = np.column_stack((X.ravel(), Y.ravel()))

    dist = np.linalg.norm(points_grid - [center_x, center_y], axis=1)
    points_grid = points_grid[dist >= radius]

    theta = np.linspace(0, 2*np.pi, n_theta, endpoint=False)
    circle_points = np.column_stack((
        center_x + radius * np.cos(theta),
        center_y + radius * np.sin(theta)
    ))

    return np.vstack((points_grid, circle_points))
```

```python

# Parmetros
L, H, D = 10.0, 5.0, 2.0
nx, ny = 60, 30
n_theta = 50

points = generate_plate_with_hole(L, H, D, nx, ny, n_theta)

# Filtro de dominio (opcional)
def domain_filter(centroid):
    center = np.array([L/2, H/2])
    return np.linalg.norm(centroid - center) >= D/2 - 1e-8

tess = Tessellation(points, domain_filter=domain_filter)

# Visualizacin
tess.plot(show_node_ids=False, show_element_ids=False, node_size=1, figure_size=(10,6))

# Identificacin de bordes
tol = 1e-6
left_nodes   = np.where(np.abs(tess.points[:,0] - 0.0) < tol)[0]
right_nodes  = np.where(np.abs(tess.points[:,0] - L)   < tol)[0]
center = np.array([L/2, H/2])
radius = D/2
dist = np.linalg.norm(tess.points - center, axis=1)
circle_nodes = np.where(np.abs(dist - radius) < 1e-4)[0]

print(f"Nodos totales: {tess.n_nodes}")
print(f"Borde izquierdo: {len(left_nodes)} nodos")
print(f"Borde derecho: {len(right_nodes)} nodos")
print(f"Borde del crculo: {len(circle_nodes)} nodos")
```

```python
elements = []
for tri in tess.simplices:
    elem = TriangularElement(
        node_ids=tri,
        nodes=tess.points,
        E = 210e9,
        nu = 0.3,
        thickness = 0.03,
        element_type="CST"
    )
    elements.append(elem)
```

```python
problem = FEMProblem(elements)
problem.assemble_stiffness()

# Nodos del extremo izquierdo
tol = 1e-8

# Fijar desplazamientos nulos
dirichlet_values = np.zeros((len(left_nodes), 2))
problem.apply_dirichlet_bcs(left_nodes, dirichlet_values)

distributed_load = [
    {
        "nodes": right_nodes,
        "angle": 0.0,      # hacia arriba
        "value": 100000.0   # N/m
    }
]

problem.apply_distributed_loads(distributed_load, tess.points)
```

```python
problem.solve(method = "cg", tol = 1e-10)
problem.compute_element_stresses()
```

```python
plot_deformed_contour(
    tess.points,
    tess.simplices,
    problem.ux,
    problem.uy,
    scale = 6000
)
```

```python
problem.plot_stiffness_sparsity(figsize=(5,5))
```

```python
problem.plot_elemental_variable(
    tess.points,
    tess.simplices,
    problem.sigma_vm / 1e6,
    title="Tensin de von Mises (MPa)"
)
```

## Problema 2

- Empotrado a la izquierda y derecha
- Carga distribuida en el orificio

```python
problem2 = FEMProblem(elements)
problem2.assemble_stiffness()

# Nodos del extremo izquierdo
tol = 1e-8

# Fijar desplazamientos nulos
dirichlet_values = np.zeros((len(left_nodes), 2))
problem2.apply_dirichlet_bcs(left_nodes, dirichlet_values)

dirichlet_values = np.zeros((len(right_nodes), 2))
problem2.apply_dirichlet_bcs(right_nodes, dirichlet_values)

distributed_load = [
    {
        "nodes": circle_nodes,
        "angle": 90,      # hacia arriba
        "value": 120000.0   # N/m
    }
]

problem2.apply_distributed_loads(distributed_load, tess.points)
```

```python
problem2.solve(method = "cg", tol = 1e-10)
problem2.compute_element_stresses()
```

```python
plot_deformed_contour(
    tess.points,
    tess.simplices,
    problem2.ux,
    problem2.uy,
    scale = 1500
)
```

```python
problem2.plot_elemental_variable(
    tess.points,
    tess.simplices,
    problem2.sigma_vm / 1e6,
    title="Tensin de von Mises (MPa)"
)
```
