# Test: Aplicación de cargas distribuidas en una placa cuadrada

En este tutorial mostraremos cómo aplicar una **carga distribuida** sobre un borde de la malla. Utilizaremos la misma placa cuadrada del tutorial anterior, empotrada en el borde inferior, pero ahora sometida a una carga uniformemente distribuida a lo largo del borde superior.

* El **patrón de esparcidad** de la matriz de rigidez global.
* La distribución de la **tensión de von Mises** en los elementos.

## 1. Descripción del problema

* **Geometría**: placa cuadrada de lado $0.156\,\text{m}$ (igual que en el Tutorial 3).
* **Material**: acero ($E = 210\,\text{GPa}$, $\nu = 0.3$), espesor $t = 1.0\,\text{m}$.
* **Condiciones de borde**: empotramiento ($u_x = u_y = 0$) en los nodos del borde inferior ($y = 0$).
* **Carga**: fuerza distribuida vertical aplicada en el borde superior ($y = 0.156\,\text{m}$), con una intensidad de $100\,000\,\text{N/m}$ (es decir, $100\,\text{kN}$ por metro lineal a lo largo del borde). La carga actúa **perpendicular al borde**, en dirección vertical. En este ejemplo elegiremos el ángulo de $90^\circ$ (hacia arriba) para ilustrar el efecto; si se deseara hacia abajo bastaría con usar $-90^\circ$.

## 2. Generación de la malla

Creamos la misma malla estructurada de $8\times8$ puntos que en el tutorial anterior. Esto nos permitirá reutilizar la numeración de nodos y facilitar la identificación del borde superior.



```python
import numpy as np
import matplotlib.pyplot as plt
from femppp import *

# Dimensiones
xl = yl = 0.156
nx = ny = 8

# Malla
x = np.linspace(0.0, xl, nx)
y = np.linspace(0.0, yl, ny)
X, Y = np.meshgrid(x, y)
points = np.column_stack((X.ravel(), Y.ravel()))

tess = Tessellation(points)
print(f"Nodos: {tess.n_nodes}, Elementos: {tess.n_elements}")
```


## 3. Creación de los elementos finitos

Creamos un objeto `TriangularElement` para cada triángulo de la malla.


```python
elements = []
for tri in tess.simplices:
    elem = TriangularElement(
        node_ids=tri,
        nodes=tess.points,
        E=210e9,
        nu=0.3,
        thickness=1.0,
        element_type="CST"
    )
    elements.append(elem)
```


## 4. Ensamblaje y condiciones de contorno

Construimos el problema, ensamblamos la matriz de rigidez y aplicamos el empotramiento en el borde inferior (nodos 0 a 7).


```python
problem = FEMProblem(elements)
problem.assemble_stiffness()

# Empotramiento inferior
dirichlet_nodes = list(range(8))
dirichlet_values = np.zeros((8, 2))
problem.apply_dirichlet_bcs(dirichlet_nodes, dirichlet_values)
```

## 5. Aplicación de la carga distribuida en el borde superior

El método `apply_distributed_loads` recibe una lista de diccionarios. Cada diccionario debe contener:

* `"nodes"`: lista de identificadores de nodos que definen el borde (en orden, de un extremo al otro).
* `"angle"`: ángulo en grados que forma la dirección de la carga con el eje $x$ positivo.  
  * $0^\circ$: dirección $+x$  
  * $90^\circ$: dirección $+y$ (hacia arriba)  
  * $-90^\circ$: dirección $-y$ (hacia abajo)
* `"value"`: magnitud de la carga en $\text{N/m}$ (fuerza por unidad de longitud a lo largo del borde).

En nuestra malla, los nodos del borde superior son aquellos con coordenada $y \approx yl$. Podemos obtenerlos mediante una condición de proximidad.


```python
# Nodos del borde superior (y ≈ yl)
top_nodes = np.where(np.isclose(tess.points[:, 1], yl))[0]
print("Nodos del borde superior:", top_nodes)
```


El vector `top_nodes` contendrá los índices en orden creciente de $x$ (de izquierda a derecha), que es justamente el orden requerido.

Ahora definimos la carga distribuida vertical hacia arriba (ángulo 90°) con intensidad $100\,000\,\text{N/m}$.


```python
distributed_loads = [
    {
        "nodes": top_nodes,
        "angle": 90.0,      # hacia arriba
        "value": 100000.0   # N/m
    }
]

problem.apply_distributed_loads(distributed_loads, tess.points)
```


> **Nota:** Si se deseara una carga hacia abajo, bastaría con cambiar el ángulo a $-90.0$.

## 6. Solución del sistema

Resolvemos mediante gradiente conjugado y calculamos las tensiones en los elementos.


```python
problem.solve(method="cg", tol=1e-10)
problem.compute_element_stresses()
```


## 7. Visualización de resultados

### 7.1 Magnitud del desplazamiento

Mostramos el campo de desplazamiento total $\overline{U} = \sqrt{u_x^2 + u_y^2}$.


```python
plot_fem_contour(
    points=tess.points,
    simplices=tess.simplices,
    nodal_values=problem.umag,
    title="Magnitud del desplazamiento (m)",
    cmap="jet",
    show_scale=True
)
```


### 7.2 Malla deformada (amplificada)

Para apreciar la deformación, amplificamos los desplazamientos con un factor de escala elevado (en este caso $500\,000$).


```python
plot_deformed_contour(
    tess.points,
    tess.simplices,
    problem.ux,
    problem.uy,
    scale=500000
)
```


### 7.3 Patrón de esparcidad de la matriz de rigidez

Podemos visualizar la estructura de la matriz de rigidez global. Esto es útil para comprender la conectividad entre grados de libertad y verificar que el ensamblaje se realizó correctamente.


```python
problem.plot_stiffness_sparsity(figsize=(5,5))
```


La matriz resultante es simétrica y presenta un patrón de bandas característico de mallas estructuradas.

### 7.4 Tensión de von Mises en los elementos

Finalmente, representamos la tensión de von Mises calculada en cada elemento (valor constante por elemento). Esta magnitud es útil para identificar las zonas más solicitadas.


```python
problem.plot_elemental_variable(
    tess.points,
    tess.simplices,
    problem.sigma_vm,
    title="Tensión de von Mises (Pa)"
)
```
